```python
import matplotlib.pyplot as plt

# Données de salaires fixes (en euros) par département
data_it = [72000, 75000, 68000, 70000, 71000, 69000, 73000, 76000, 67000, 74000]
data_sales = [58000, 60000, 62000, 59000, 61000, 57000, 63000, 65000, 55000, 64000]
data_hr = [50000, 52000, 54000, 51000, 53000, 49000, 55000, 56000, 48000, 57000]

# Regroupement des données
data_salaires = [data_it, data_sales, data_hr]
departements = ['IT', 'Ventes', 'RH']

# Création du boxplot
box = plt.boxplot(data_salaires, labels=departements, patch_artist=True, showmeans=True)

# Appliquer des couleurs personnalisées
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Titre et légendes
plt.title('Comparaison des Salaires par Département')
plt.ylabel('Salaire Annuel (€)')
plt.grid(axis='y', alpha=0.5)

# Affichage
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Données fixes
data_it = [72000, 75000, 68000, 70000, 71000, 69000, 73000, 76000, 67000, 74000]
data_sales = [58000, 60000, 62000, 59000, 61000, 57000, 63000, 65000, 55000, 64000]
data_hr = [50000, 52000, 54000, 51000, 53000, 49000, 55000, 56000, 48000, 57000]

# Création d’un DataFrame long pour Seaborn
df = pd.DataFrame({
    'Salaire': data_it + data_sales + data_hr,
    'Département': ['IT'] * len(data_it) + ['Ventes'] * len(data_sales) + ['RH'] * len(data_hr)
})

# Création du boxplot avec Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x='Département', y='Salaire', data=df, palette='pastel', showmeans=True,
            meanprops={"marker": "o", "markerfacecolor": "white", "markeredgecolor": "black"})

# Titres et habillage
plt.title('Comparaison des Salaires par Département (Seaborn)', fontsize=14)
plt.ylabel('Salaire Annuel (€)')
plt.grid(axis='y', alpha=0.5)
plt.show()
```
