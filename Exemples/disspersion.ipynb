Analyse de la relation entre l'âge d'un employé et son salaire annuel au sein d'une entreprise. On pourrait aussi ajouter une troisième dimension via la couleur (par exemple, le département) pour voir si des tendances salariales diffèrent par département.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Générer des données réalistes
np.random.seed(1)
n = 100
age = np.random.normal(35, 8, n).astype(int)  # moyenne d’âge autour de 35 ans
departements = np.random.choice(['IT', 'RH', 'Finance', 'Marketing', 'Ventes'], n)

# Salaires réalistes selon département
salaire = []
for i in range(n):
    base = 28000
    if departements[i] == 'IT':
        base = 40000
    elif departements[i] == 'Finance':
        base = 38000
    elif departements[i] == 'Marketing':
        base = 35000
    elif departements[i] == 'Ventes':
        base = 36000
    elif departements[i] == 'RH':
        base = 32000

    montant = base + age[i]*500 + np.random.normal(0, 4000)
    salaire.append(montant)

# Créer un DataFrame
df = pd.DataFrame({
    'Âge': age,
    'Salaire (€)': salaire,
    'Département': departements
})

# Couleurs simples par département
departement_colors = {
    'IT': 'blue',
    'RH': 'green',
    'Finance': 'orange',
    'Marketing': 'red',
    'Ventes': 'purple'
}

```

```python
plt.figure(figsize=(10, 6))
for dept in df['Département'].unique():
    subset = df[df['Département'] == dept]
    plt.scatter(subset['Âge'], subset['Salaire (€)'],
                label=dept, color=departement_colors[dept], alpha=0.7)

plt.title('Relation Âge vs Salaire par Département (Matplotlib)', fontsize=14)
plt.xlabel('Âge (ans)')
plt.ylabel('Salaire annuel (€)')
plt.grid(True)
plt.legend(title='Département')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Âge', y='Salaire (€)', hue='Département', palette='Set2', s=90, alpha=0.8)
plt.title("Relation Âge vs Salaire par Département (Seaborn)", fontsize=14)
plt.xlabel("Âge (ans)")
plt.ylabel("Salaire annuel (FG)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
