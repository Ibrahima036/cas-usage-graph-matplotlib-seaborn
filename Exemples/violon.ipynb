Exemple concret : Comparer la distribution des prix de logements par quartier, pour voir non seulement le prix médian mais aussi la concentration des prix dans chaque quartier.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Exemple de 4 quartiers
quartiers = ['Centre-ville', 'Banlieue', 'Quartier Nord', 'Quartier Sud']

# Génération de prix réalistes (en milliers d'euros)
np.random.seed(42)
prix = np.concatenate([
    np.random.normal(350, 30, 50),    # Centre-ville
    np.random.normal(200, 40, 50),    # Banlieue
    np.random.normal(270, 50, 50),    # Quartier Nord
    np.random.normal(220, 20, 50)     # Quartier Sud
])

quartier_labels = quartiers[0]*50 + quartiers[1]*50 + quartiers[2]*50 + quartiers[3]*50
df = pd.DataFrame({
    'quartier': quartiers * 50,
    'prix_logement': prix
})
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='quartier', y='prix_logement', data=df, palette='pastel', inner='box')
plt.title('Distribution des prix de logements par quartier')
plt.xlabel('Quartier')
plt.ylabel('Prix du logement (en FG)')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
