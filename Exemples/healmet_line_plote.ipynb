Exemple concret : Matrice de corrélation des caractéristiques d'un dataset pour identifier les relations entre elles.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Création d’un DataFrame réaliste sur des logements
data = {
    'surface_m2': [35, 45, 60, 80, 100, 120, 150, 200, 250, 300],
    'nombre_pieces': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'etage': [0, 1, 2, 2, 3, 3, 4, 5, 6, 7],
    'prix_k€': [100, 120, 180, 220, 300, 350, 420, 520, 600, 720]
}

df = pd.DataFrame(data)
```

```python
# Matrice de corrélation
correlation = df.corr(numeric_only=True)

# Affichage de la heatmap avec Matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Coefficient de corrélation')

# Ajout des noms de colonnes
tick_labels = correlation.columns
plt.xticks(ticks=np.arange(len(tick_labels)), labels=tick_labels, rotation=45, ha='right')
plt.yticks(ticks=np.arange(len(tick_labels)), labels=tick_labels)

# Affichage des coefficients dans la matrice
for i in range(len(correlation.columns)):
    for j in range(len(correlation.columns)):
        value = round(correlation.iloc[i, j], 2)
        plt.text(j, i, f'{value}', ha='center', va='center', color='black')

plt.title('Matrice de corrélation')
plt.tight_layout()
plt.show()
```

Visualisation de l'évolution du nombre d'utilisateurs actifs mensuels (MAU) pour une application mobile sur les 12 derniers mois, avec une distinction entre utilisateurs iOS et Android.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données sur 12 mois
data = {
    'mois': ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin', 'Juil', 'Août', 'Sep', 'Oct', 'Nov', 'Déc'],
    'Android': [1200, 1350, 1500, 1600, 1700, 1750, 1900, 2000, 2100, 2200, 2300, 2500],
    'iOS':     [800, 850, 900, 950, 1100, 1150, 1200, 1250, 1300, 1400, 1450, 1600]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['mois'], df['Android'], marker='o', label='Android', color='green')
plt.plot(df['mois'], df['iOS'], marker='o', label='iOS', color='blue')

plt.title('Évolution mensuelle des utilisateurs actifs (MAU)')
plt.xlabel('Mois')
plt.ylabel('Nombre d’utilisateurs actifs')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Restructurer les données pour Seaborn (long format)
df_long = df.melt(id_vars='mois', var_name='Plateforme', value_name='Utilisateurs')

plt.figure(figsize=(10,6))
sns.lineplot(data=df_long, x='mois', y='Utilisateurs', hue='Plateforme', marker='o', palette='Set2')

plt.title('MAU par plateforme (Android vs iOS)')
plt.xlabel('Mois')
plt.ylabel('Utilisateurs actifs mensuels')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
