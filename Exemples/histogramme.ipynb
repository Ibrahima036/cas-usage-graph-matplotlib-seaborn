Distribution des âges des clients dans une base de données, pour déterminer si la clientèle est majoritairement jeune, âgée, ou uniformément répartie.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données fictives mais réalistes
ages_clients = [22, 25, 27, 30, 33, 35, 36, 40, 42, 45, 47, 50, 52, 55, 58, 60, 62, 65, 68, 70,
                23, 26, 28, 31, 34, 37, 38, 41, 43, 46, 48, 51, 53, 56, 59, 61, 63, 66, 69, 72,
                24, 29, 32, 39, 44, 49, 54, 57, 64, 67,
                21, 20, 22, 23, 25, 27, 30, 31, 35, 36,
                38, 40, 42, 45, 47, 50, 53, 55, 58, 60,
                62, 65, 67, 69, 70, 72, 75, 77, 79, 80,
                26, 28, 33, 34, 37, 41, 46, 48, 51, 59]

df = pd.DataFrame({'age': ages_clients})

```

```python
#Matplotlib
plt.figure(figsize=(10,6))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution des âges des clients')
plt.xlabel('Âge')
plt.ylabel('Nombre de clients')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Seaborn
plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=10, kde=True, color='salmon', edgecolor='black')
plt.title('Distribution des âges des clients (avec Seaborn)')
plt.xlabel('Âge')
plt.ylabel('Nombre de clients')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
